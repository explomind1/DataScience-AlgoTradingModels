```python


def chunks(lst,n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
        
```

```python
def isSupport(df,stock,i):
  support = df['Low'][stock][i] < df['Low'][stock][i-1]  and df['Low'][stock][i] < df['Low'][stock][i+1] \
  and df['Low'][stock][i+1] < df['Low'][stock][i+2] and df['Low'][stock][i-1] < df['Low'][stock][i-2]

  return support

def isResistance(df,stock,i):
  resistance = df['High'][i] > df['High'][stock][i-1]  and df['High'][stock][i] > df['High'][stock][i+1] \
  and df['High'][stock][i+1] > df['High'][stock][i+2] and df['High'][stock][i-1] > df['High'][stock][i-2] 

  return resistance
```
